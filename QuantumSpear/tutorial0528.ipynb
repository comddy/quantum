```python
from QuantumFinance import QuantumFinance
import numpy as np
```

```python
#预期收益率
return_rate = np.array([ 0.01528439, -0.00078095,  0.00051792,  0.00087001])
#风险协方差矩阵
risk_matrix = np.array([[ 2.54138859e-03,  7.34022167e-05,  1.28600531e-04,-9.98612132e-05],
                        [ 7.34022167e-05,  2.58486713e-04,  5.30427595e-05,4.44816208e-05],
                        [ 1.28600531e-04,  5.30427595e-05,  7.91504681e-04,-1.23887382e-04],
                        [-9.98612132e-05,  4.44816208e-05, -1.23887382e-04,1.97892585e-04]])
```

```python
qf = QuantumFinance(return_rate=return_rate,risk_matrix=risk_matrix,num_assets=4,global_seed=10,risk_factor=0.5,g=2)
```

```python
qf.solve()
```
